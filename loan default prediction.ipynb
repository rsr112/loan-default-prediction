```python
from scipy.stats import chi2_contingency
def chi_square_test(data):
    stat, p, dof, expected = chi2_contingency(data)
    alpha = 0.05
    print("p value is " + str(p))
    if p <= alpha:
        print('Dependent (reject H0)')
    else:
        print('Independent (H0 holds true)')
```

```python
car_ownership_risk_flag = pd.crosstab(loan["Car_Ownership"], loan["Risk_Flag"])
car_ownership_risk_flag
```

```python
chi_square_test(car_ownership_risk_flag)
```

```python
marital_status_risk_flag = pd.crosstab(loan["Married/Single"], loan["Risk_Flag"])
marital_status_risk_flag
```

```python
chi_square_test(marital_status_risk_flag)
```

```python
house_ownership_risk_flag = pd.crosstab(loan["House_Ownership"], loan["Risk_Flag"])
house_ownership_risk_flag
```

```python
chi_square_test(house_ownership_risk_flag)
```

```python
features = ["CURRENT_JOB_YRS","Experience"]

loan_for_pca = loan[features]
scaled_loan_for_pca = (loan_for_pca - loan_for_pca.mean(axis=0))/loan_for_pca.std()
scaled_loan_for_pca
```

```python
from sklearn.decomposition import PCA
pca = PCA()
loan_pca = pca.fit_transform(scaled_loan_for_pca)
component_names = [f"PC{i+1}" for i in range(loan_pca.shape[1])]
loan_pca = pd.DataFrame(loan_pca, columns=component_names)

loan_pca.head()
```

```python
loan1 = pd.concat([loan,loan_pca],axis=1)
loan1.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
house_ownership_encoded = onehot_encoder.fit_transform(loan1['House_Ownership'].values.reshape(-1, 1))
house_ownership_loan1 = pd.DataFrame(house_ownership_encoded, 
                                   columns=onehot_encoder.get_feature_names_out(['House_Ownership']))
loan1 = loan1.drop('House_Ownership', axis=1)
loan1 = pd.concat([loan1, house_ownership_loan1], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for col in ['Married/Single','Car_Ownership']:
    loan1[col] = label_encoder.fit_transform(loan1[col])
```

```python
high_card_features = ['Profession', 'CITY', 'STATE']

for col in high_card_features:
    value_counts = loan1[col].value_counts()
    loan1[col + '_count'] = loan1[col].map(value_counts)
```

```python
from sklearn.preprocessing import OrdinalEncoder
cat_cols = ['Profession','CITY','STATE']
encoder = OrdinalEncoder()
loan1[cat_cols] = encoder.fit_transform(loan1[cat_cols])
```

```python
from sklearn.model_selection import train_test_split
X = loan1.drop(['Risk_Flag'],axis=1)
y = loan1.Risk_Flag
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 500)
X_res, y_res = sm.fit_resample(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter = 500000)
model.fit(X_res, y_res)
y_pred = model.predict(X_test)
accuracy = model.score(X_test, y_test)
accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion="entropy",random_state=420)
model.fit(X_res, y_res)
y_pred = model.predict(X_test)
accuracy = model.score(X_test, y_test)
accuracy
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(criterion='gini', bootstrap=True, random_state=420)
model.fit(X_res, y_res)
y_pred = model.predict(X_test)
accuracy = model.score(X_test, y_test)
accuracy
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
model.fit(X_res, y_res)
y_pred = model.predict(X_test)
accuracy = model.score(X_test, y_test)
accuracy
```

```python
print(classification_report(y_test,y_pred))
```
